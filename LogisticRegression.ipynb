```python
import os
import numpy as np
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import seaborn as sns
df = sns.load_dataset("titanic")
import seaborn as sns

#Load the dataset
df = sns.load_dataset("titanic")
```

```python
#data exploration, data preparation

#Get summary of data

print(df.isna().sum())
```

```python
#Define features and target
features = ['Sex', 'Age', 'embark_Town']
target = ['survived']
```

```python
df.head()

print(df.isnull().sum())

#Handle missing values
titanicModel = df.dropna()
df.head()

print(df.isnull().sum())
titanicModel = df.dropna()
print(df)

titanicModel = pd.get_dummies(titanicModel, columns=['Sex', 'Age', 'embark_Town'], drop_first=True)
titanicModel.head()
```

```python
#Define X and y
X = titanicModel['Sex', 'Age', 'embark_town']
y = titanicModel[target]

#Display head of X and y
X.head(6)
y.head(6)
```

```python
print(df.isnull().sum())
titanicModel = df.dropna()
print(df)
titanicModel.head()
```

```python
#Train the model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

y_pred

titanic_df = pd.DataFrame({
    "Sex" : X_test["Sex"].values,
    "" : y_test.values,
    "" : y_pred
})

titanic_df.head()
```
